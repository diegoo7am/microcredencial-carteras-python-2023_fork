# Gestión riesgos de las carteras


**Desde la crisis financiera global en 2008, la gestión de riesgos se ha vuelto particularmente más importante que la gestión del rendimiento en la optimización de carteras.**

La gestión del **riesgo de cartera** es el proceso de identificar, evaluar y mitigar los diversos riesgos asociados con una cartera de inversiones.

Implica implementar estrategias para optimizar el equilibrio entre riesgo y retorno, asegurando que la cartera se alinee con los **objetivos financieros y la tolerancia al riesgo del inversor**. 

Una gestión de riesgos efectiva ayuda a minimizar las pérdidas potenciales, optimizar los retornos y mejorar el rendimiento de la inversión a largo plazo.

### Componentes clave de la gestión del riesgo de cartera:

1. **Identificación de Riesgos:** Implica reconocer los diversos riesgos asociados con una cartera de inversiones, incluidos los relacionados con activos específicos, sectores o estrategias de inversión.

2. **Evaluación y Medición de Riesgos:** Involucra cuantificar el impacto potencial de los riesgos identificados en el desempeño de la cartera de inversiones. Este paso utiliza diversas herramientas y métricas, como el valor en riesgo (VaR), la desviación estándar, la beta etc.

3. **Estrategias de Mitigación de Riesgos:** Implica implementar tácticas para reducir el impacto potencial de los riesgos identificados en la cartera. Estas estrategias pueden incluir diversificación, cobertura y rebalanceo.

4. **Monitorización y Revisión de Riesgos:** Involucra la evaluación continua y la valoración del perfil de riesgo de la cartera y la efectividad de las estrategias de gestión de riesgos implementadas. Las revisiones regulares de la cartera ayudan a los inversores a tomar decisiones informadas y ajustar sus estrategias según sea necesario.


### Estrategias clave para la gestión del riesgo en carteras:

- **Diversificación:** Repartir inversiones en diferentes clases de activos y regiones para minimizar el impacto de riesgos específicos.

- **Rebalanceo:** Ajustar la asignación de activos para mantener el nivel de riesgo deseado.
  
- **Cobertura:** Usar instrumentos financieros, por ejemplo uso de derivaods, para reducir o compensar el riesgo de movimientos de precios adversos. 

- **Retornos Ajustados por Riesgo:** Evaluar el rendimiento considerando el retorno y el riesgo, para tomar decisiones informadas.
  
- **Pruebas de Estrés y Análisis de Escenarios:** Evaluar el impacto de condiciones de mercado adversas en la cartera para identificar y ajustar vulnerabilidades.


Estas estrategias ayudan a optimizar el equilibrio entre riesgo y retorno, protegiendo y potenciando el valor de la cartera a largo plazo.


En este notebook vamos a analizar distintas maneras de medir el riesgo de un activo o cartera.

La medición del riesgo de un activo o cartera se enfoca en analizar las colas de la distribución de retornos, destacando los resultados extremos tanto positivos como negativos.

Este análisis permite evaluar la probabilidad e impacto de eventos de mercado inusuales, más allá de lo que indican medidas tradicionales como la volatilidad.

Al entender estos extremos, los inversores pueden diseñar estrategias de mitigación más efectivas, preparándose para escenarios de alto riesgo y protegiendo mejor sus inversiones contra sorpresas adversas.

Aparte de las medidas de riesgo ya estudiadas, las principales medidas o ratios son:


### 1- Valor en Riesgo (VaR)

El Valor en Riesgo (VaR) es una técnica de gestión de riesgos financieros que mide y cuantifica el nivel de riesgo financiero dentro de una empresa, portafolio o posición durante un período específico. El VaR indica la máxima pérdida esperada con un nivel de confianza dado, por ejemplo, un 95% o 99%, durante el período de tiempo especificado.

#### Definición

El VaR se puede definir formalmente como:

$$
\text{VaR}_{\alpha}(X) = -\inf\{x \in \mathbb{R} : P(X \leq x) \geq \alpha\}
$$

Donde:
- $X$ representa las pérdidas y ganancias.
- $\alpha$ es el nivel de confianza (por ejemplo, 0.95 para el 95% de confianza).
- $P$ es la función de probabilidad.
- $\inf$ denota el ínfimo, proporcionando el valor más bajo de $x$ para el cual la probabilidad acumulada excede el nivel de confianza $\alpha$.

#### Métodos de Cálculo

Existen varios métodos para calcular el VaR, incluyendo:

1. **Método Paramétrico (Varianza-Covarianza)**: Asume que los retornos de los activos siguen una distribución normal y utiliza la varianza y la covarianza de los retornos para estimar el VaR.

2. **Método Histórico**: Calcula el VaR directamente de la distribución histórica de rendimientos, sin asumir una distribución específica.

3. **Método de Simulación de Montecarlo**: Utiliza simulación para generar una amplia gama de posibles resultados futuros y sus probabilidades asociadas para calcular el VaR.

#### Aplicación

El VaR se utiliza ampliamente en la industria financiera para la gestión de riesgos, la asignación de capital y la regulación financiera. Proporciona una medida clara y concisa del riesgo máximo que una cartera puede enfrentar, lo que ayuda a los inversores y a los gestores de riesgo a tomar decisiones informadas sobre sus estrategias de inversión y gestión de riesgos.

### 2-  Valor en Riesgo Condicional (CVaR)

El Valor en Riesgo Condicional (CVaR), también conocido como Riesgo de Pérdida Esperada o Media en Riesgo (Expected Shortfall o Mean Shortfall), es una medida de riesgo financiero que proporciona una estimación del riesgo de cola o el riesgo de pérdidas extremas. A diferencia del Valor en Riesgo (VaR), que ofrece un límite superior de pérdidas para un nivel de confianza específico, el CVaR calcula el promedio de las pérdidas que exceden el VaR, ofreciendo una medida más completa del riesgo de eventos extremos.

#### Definición

Formalmente, el CVaR para un nivel de confianza $\alpha$ se define como el promedio de las pérdidas que superan el VaR en ese nivel de confianza:

$$
\text{CVaR}_\alpha = -\frac{1}{1-\alpha} \int_{\alpha}^{1} \text{VaR}_u \, du
$$

Donde:
- $\text{VaR}_u$ es el Valor en Riesgo al nivel de confianza $u$.
- $\alpha$ es el nivel de confianza (por ejemplo, 0.95 para el 95% de confianza).

#### Métodos de Cálculo

Al igual que el VaR, el CVaR puede ser calculado usando varios métodos, incluyendo el método histórico, la simulación de Montecarlo y el método paramétrico. El cálculo del CVaR típicamente sigue el cálculo del VaR, y se enfoca en la porción de la distribución de retornos más allá del VaR.

#### Aplicación

El CVaR es ampliamente utilizado en la gestión de riesgos financieros para evaluar y controlar el riesgo de pérdidas extremas. Al proporcionar una medida del riesgo esperado en el peor de los casos (más allá del VaR), el CVaR ayuda a los inversores y gestores de riesgos a comprender mejor el riesgo de cola y a prepararse para posibles eventos de mercado adversos. Es especialmente valioso en la evaluación de carteras y estrategias de inversión de alto riesgo, donde la gestión del riesgo de eventos extremos es crítica.

#### Ventajas sobre el VaR

- **Mayor Sensibilidad al Riesgo de Cola**: Al enfocarse en las pérdidas más allá del VaR, el CVaR captura el riesgo de eventos extremadamente adversos, que el VaR por sí solo puede subestimar.
- **Coherencia del Riesgo**: A diferencia del VaR, el CVaR satisface la propiedad de coherencia del riesgo, lo que significa que refleja adecuadamente la diversificación y no incentiva la concentración de riesgo.


### 3- Máximo Drawdown (Max Drawdown)

El Máximo Drawdown (Max Drawdown) es una medida de riesgo que indica la mayor caída desde un pico hasta un valle en el valor de una cartera o activo, antes de un nuevo pico, durante un período específico. Esta métrica es especialmente útil para determinar la tolerancia al riesgo y la potencial pérdida máxima que un inversor podría enfrentar.

#### Definición

El Max Drawdown se calcula como la máxima diferencia negativa entre un pico y el valle subsecuente en el valor de la cartera:

$$
\text{Max Drawdown} = \max_{t \in [0, T]} \left( \max_{t' \in [0, t]} (V_{t'}) - V_t \right)
$$

Donde:
- $T$ es el período total de tiempo considerado.
- $V_t$ es el valor de la cartera o activo en el tiempo $t$.
- $\max_{t' \in [0, t]} (V_{t'})$ es el valor máximo de la cartera hasta el tiempo $t$.

#### Cálculo

1. Identificar todos los picos (máximos locales) en el gráfico del valor de la cartera a lo largo del tiempo.
2. Para cada pico, calcular la caída hasta el valle más bajo subsecuente (mínimo local) antes del próximo pico.
3. El Max Drawdown es la mayor de estas caídas.


#### Aplicación

El Max Drawdown es crítico para la evaluación del riesgo en la gestión de inversiones. Proporciona a los inversores una medida de la pérdida potencial máxima y el riesgo de caídas significativas en el valor de la cartera. Es particularmente valioso para la planificación financiera y la estrategia de inversión, ayudando a los inversores a entender la volatilidad y el riesgo de sus inversiones.

#### Ventajas

- **Medida Intuitiva de Riesgo**: Ofrece una perspectiva clara de las pérdidas potenciales que podrían experimentarse durante períodos de declive del mercado.
- **Relevancia para la Toma de Decisiones**: Ayuda a los inversores a evaluar si el nivel de riesgo de una inversión es aceptable en relación con sus objetivos y tolerancia al riesgo.

El Max Drawdown es una herramienta esencial en la caja de herramientas de cualquier inversor, proporcionando una visión crítica del peor escenario posible, lo que es fundamental para la toma de decisiones informadas sobre riesgos y recompensas.


### 4- Tiempo de Recuperación Medio (Average Recovery Time)

El Tiempo de Recuperación Medio es una métrica financiera que indica, en promedio, cuánto tiempo toma para que una inversión o cartera se recupere de sus caídas durante un período específico. A diferencia del Tiempo de Recuperación, que se enfoca en un solo evento de drawdown y recuperación, el Tiempo de Recuperación Medio considera múltiples eventos de caída y recuperación para proporcionar una medida agregada y representativa del comportamiento de recuperación de la inversión.

#### Definición

El Tiempo de Recuperación Medio se calcula como el promedio de los tiempos de recuperación de todos los drawdowns identificados en el período de análisis.

#### Cálculo

1. Identificar todos los eventos de drawdown durante el período de análisis, marcando el inicio (pico previo) y el fin (recuperación al valor del pico) de cada evento.
2. Calcular el Tiempo de Recuperación para cada evento de drawdown individual, que es el número de periodos desde el inicio del drawdown hasta la recuperación completa.
3. Sumar todos los tiempos de recuperación calculados y dividir por el número total de eventos de drawdown para obtener el promedio.

#### Fórmula

$$
\text{Tiempo de Recuperación Medio} = \frac{\sum_{i=1}^{n} \text{Tiempo de Recuperación}_i}{n}
$$

Donde:
- $n$ es el número total de eventos de drawdown.
- $\text{Tiempo de Recuperación}_i$ es el tiempo de recuperación para el $i$-ésimo evento de drawdown.

#### Aplicación

El Tiempo de Recuperación Medio es útil para inversores y gestores de fondos al evaluar la resiliencia de una inversión y su capacidad para recuperarse de períodos adversos. Una métrica menor indica una inversión que se recupera rápidamente de las caídas, potencialmente haciéndola más atractiva para aquellos con menor tolerancia al riesgo o necesidades financieras a corto plazo.

#### Consideraciones

- **Estrategia de Inversión**: Inversiones con tiempos de recuperación medios cortos pueden ser preferibles para estrategias de inversión que buscan minimizar el impacto negativo de las volatilidades del mercado.
- **Planificación Financiera**: Comprender el Tiempo de Recuperación Medio ayuda a los inversores a planificar mejor sus necesidades financieras, alineando sus inversiones con sus horizontes temporales y tolerancia al riesgo.

Incluir el análisis del Tiempo de Recuperación Medio junto con otras métricas de riesgo ofrece una vista holística del perfil de riesgo y rendimiento de una inversión, facilitando la toma de decisiones informadas y estratégicas.

### 5- Ratio de Información

El Ratio de Información es una medida de rendimiento relativo que evalúa el exceso de retorno de un portafolio sobre un benchmark o índice de referencia, ajustado por el riesgo de sobre o subrendimiento. Es decir, mide la habilidad de un gestor de portafolio para generar retornos adicionales por encima de un índice de referencia dado, ajustado por la volatilidad de esos excesos de retorno. Se considera una herramienta valiosa para evaluar gestores de fondos y estrategias de inversión. La fórmula para calcular el Ratio de Información es la siguiente:

#### Fórmula del Ratio de Información

$$
\text{Ratio de Información} = \frac{R_p - R_b}{\sigma_{(R_p - R_b)}}
$$

Donde:

- $R_p$ = Retorno promedio del portafolio
- $R_b$ = Retorno promedio del benchmark o índice de referencia
- $\sigma_{(R_p - R_b)}$ = Desviación estándar de la diferencia entre los retornos del portafolio y los retornos del benchmark

#### Interpretación

- **Valor Positivo**: Indica que el portafolio ha superado al benchmark, generando retornos adicionales por encima de los del índice de referencia, ajustados por el riesgo de esos excesos de retorno. Cuanto mayor sea el valor, mejor será la gestión del portafolio en relación con su benchmark.
- **Valor Negativo**: Sugiere que el portafolio ha rendido por debajo del benchmark, indicando una gestión inferior.
- **Valor Cero**: Implica que el portafolio ha igualado el rendimiento del benchmark.

#### Aplicación

El Ratio de Información es especialmente útil para inversores institucionales y otros profesionales de la inversión que evalúan la habilidad de los gestores de fondos para superar consistentemente un índice de referencia. Permite una comparación efectiva entre diferentes gestores y estrategias de inversión, centrándose en la generación de valor ajustado al riesgo en relación con un mercado o sector específico.


### 6- Ratio de Treynor

El Ratio de Treynor es una medida de rendimiento ajustado al riesgo que se utiliza para evaluar la eficiencia de un portafolio de inversión o un activo en comparación con el riesgo de mercado en su conjunto. A diferencia del Ratio de Sharpe, que se ajusta por la volatilidad total, el Ratio de Treynor solo considera el riesgo sistemático, reflejado en el beta del portafolio. Esto lo hace especialmente útil para inversores que poseen carteras diversificadas donde el riesgo no sistemático ha sido minimizado. La fórmula para calcular el Ratio de Treynor es la siguiente:

#### Fórmula del Ratio de Treynor

$$
\text{Ratio de Treynor} = \frac{R_p - R_f}{\beta_p}
$$

Donde:

- $R_p$ = Retorno promedio del portafolio
- $R_f$ = Tasa de retorno libre de riesgo
- $\beta_p$ = Beta del portafolio, que mide la sensibilidad de sus retornos en relación con los retornos del mercado

#### Interpretación

El Ratio de Treynor ofrece una perspectiva del exceso de rendimiento generado por cada unidad de riesgo sistemático asumido. Valores más altos indican una mayor eficiencia en la generación de retornos ajustados al riesgo, mientras que valores más bajos sugieren una eficiencia menor. Comparando los Ratios de Treynor de diferentes portafolios o activos, los inversores pueden identificar aquellos que han logrado un mejor rendimiento relativo al riesgo de mercado.

#### Aplicación

Es particularmente útil para comparar la eficiencia de diferentes portafolios o fondos que tienen diferentes niveles de exposición al mercado. Al enfocarse en el riesgo sistemático, el Ratio de Treynor es relevante para inversores que mantienen carteras diversificadas, donde el riesgo no sistemático (específico del activo) se considera mitigado.



### 7- Ratio de Sortino

El Ratio de Sortino es una variación del ratio de Sharpe que se enfoca únicamente en la volatilidad de los retornos negativos, ignorando la volatilidad de los retornos positivos. Se considera una medida más relevante para los inversores que están especialmente preocupados por las caídas en el valor de sus inversiones. La fórmula para calcular el Ratio de Sortino es la siguiente:

#### Fórmula del Ratio de Sortino

$$
\text{Ratio de Sortino} = \frac{R_p - R_f}{\sigma_d}
$$

Donde:

- $R_p$ = Retorno promedio del portafolio
- $R_f$ = Tasa de retorno libre de riesgo
- $\sigma_d$ = Desviación estándar de los retornos negativos del portafolio (riesgo de baja)

#### Cálculo de la Desviación Estándar de los Retornos Negativos ($\sigma_d$)

La desviación estándar de los retornos negativos, o el riesgo de baja, se calcula utilizando solo los retornos que están por debajo del retorno objetivo o la tasa libre de riesgo. Esto se puede expresar como:

$$
\sigma_d = \sqrt{\frac{\sum_{i=1}^{N} \min[(R_i - R_f), 0]^2}{N_{neg}}}
$$

Donde:

- $R_i$ = Retorno del portafolio en el período $i$
- $N_{neg}$ = Número de períodos con retornos negativos

El Ratio de Sortino, por lo tanto, ofrece una perspectiva sobre el rendimiento ajustado al riesgo que se centra específicamente en el riesgo de obtener retornos por debajo de un umbral deseado, en lugar de la volatilidad total del portafolio.


### 8- Ratio de Calmar

El Ratio de Calmar, desarrollado por Terry W. Calmar en 1991, es una medida de rendimiento ajustado al riesgo que se utiliza para evaluar la eficiencia de inversiones, particularmente fondos de inversión y carteras de activos. Este ratio compara el rendimiento anualizado de un portafolio con su máximo drawdown (la máxima pérdida desde un pico hasta un valle antes de un nuevo máximo) durante un período específico. Es especialmente útil para medir cómo se comporta un portafolio durante períodos de estrés del mercado. La fórmula para calcular el Ratio de Calmar es la siguiente:

#### Fórmula del Ratio de Calmar

$$
\text{Ratio de Calmar} = \frac{R_a}{MD}
$$

Donde:

- $R_a$ = Retorno anualizado del portafolio
- $MD$ = Máximo Drawdown del portafolio en el mismo período

#### Interpretación

- **Valor Alto**: Indica una relación favorable entre el retorno anualizado y el riesgo de grandes pérdidas, sugiriendo que el portafolio ofrece buenos retornos ajustados al riesgo de drawdown.
- **Valor Bajo**: Sugiere que el portafolio ha experimentado grandes pérdidas en comparación con su retorno anualizado, indicando un mayor riesgo relativo.

#### Aplicación

El Ratio de Calmar es particularmente valioso para inversores que están preocupados por las pérdidas significativas y buscan estrategias de inversión que maximicen los retornos mientras minimizan las caídas profundas. Al centrarse en el máximo drawdown como medida de riesgo, el Ratio de Calmar proporciona una perspectiva sobre la tolerancia al riesgo y la estabilidad del rendimiento de un portafolio durante períodos de declive del mercado.


### 9- Bull Beta y Bear Beta

El análisis de Bull Beta y Bear Beta proporciona una comprensión más profunda del comportamiento de un activo o portafolio durante condiciones específicas del mercado, particularmente en mercados alcistas (bull markets) y bajistas (bear markets), respectivamente. Estas métricas ajustan la idea tradicional del beta del mercado para reflejar cómo un activo responde a condiciones de mercado divergentes.

#### Bull Beta

- **Definición**: Bull Beta mide la sensibilidad de los retornos de un activo o portafolio respecto a los retornos de un índice de referencia durante períodos de mercado alcista. Un Bull Beta mayor que 1 indica que el activo tiende a superar al mercado en ascenso, mientras que un valor menor que 1 sugiere que el activo tiene un desempeño inferior en comparación con el mercado en estos períodos.

#### Bear Beta

- **Definición**: Bear Beta, por otro lado, mide la sensibilidad de los retornos de un activo o portafolio respecto a los retornos de un índice de referencia durante períodos de mercado bajista. Un Bear Beta mayor que 1 indica que el activo tiende a tener un peor desempeño que el mercado en descenso, mientras que un valor menor que 1 sugiere que el activo resiste mejor las condiciones del mercado bajista.

#### Cálculo

Para calcular Bull Beta y Bear Beta, primero se debe clasificar el rendimiento del mercado en períodos alcistas y bajistas. Luego, se calcula el coeficiente de correlación y la volatilidad relativa del activo frente al mercado para cada uno de estos períodos, utilizando la fórmula estándar de beta:

$$
\beta = \frac{Cov(R_a, R_m)}{Var(R_m)}
$$

Donde:
- $R_a$ es el retorno del activo.
- $R_m$ es el retorno del mercado.
- $Cov$ es la covarianza entre los retornos del activo y del mercado.
- $Var$ es la varianza de los retornos del mercado.

Para Bull Beta, se utiliza solo la data correspondiente a los períodos alcistas para calcular $Cov$ y $Var$, mientras que para Bear Beta se utiliza la data de los períodos bajistas.

#### Aplicación

- **Bull Beta** es útil para inversores que buscan capitalizar en mercados en ascenso identificando activos que potencialmente ofrecerán retornos superiores.
- **Bear Beta** es valioso para inversores enfocados en la preservación del capital durante períodos de declive del mercado, permitiéndoles identificar activos que posiblemente minimizarán las pérdidas.

Estas métricas permiten a los inversores ajustar sus portafolios basándose en sus expectativas del mercado y su tolerancia al riesgo, ofreciendo una herramienta más matizada que el beta tradicional para la gestión de la diversificación y el riesgo.

## VaR


La pregunta que hay que contestar es : **¿CUÁNTO PUEDO PERDER?**

El Valor en Riesgo (VaR) es una medida del riesgo de mercado de un activo o una cartera de activos. Se define como la **pérdida máxima que se espera que un activo o una cartera de activos sufra en un período de tiempo determinado con un cierto nivel de confianza**.

Por ejemplo, si el VaR diario de una acción es de 10 de euros con un nivel de confianza del 95%, esto significa que el 95% de los días perderemos menos de 10 euros.

El VaR facilita la estimación de la cantidad máxima que podría perderse en el mercado, proporcionando así una base para determinar el **capital necesario para mantener la posición**.

También permite protegerse frente a grandes pérdidas por la concentración de riesgos que resultarían inaceptables.

Para calcular el VaR existen tres métodos:

* **Método paramétrico**: Se asume que los retornos siguen una distribución normal.

* **Método históricos**: Se calcula el VaR directamente a partir de los datos históricos.

* **Método de simulación**: Se emplea las simulaciones de Montecarlo.

### VaR paramétrico

El **VaR paramétrico** es una técnica para estimar el riesgo de pérdida de una cartera de inversiones. Se basa en la suposición de que los rendimientos de los activos siguen una distribución normal.

**Una hipótesis importante es que las distribuciones son simétricas**.

Los pasos para calcular el VaR paramétrico son:

- Estimar μ y σ de los rendimientos.
  
- Elegir el nivel de confianza (ej. 95%).
  
- Usar ppf para obtener el valor correspondiente. La función ppf es la inversa de la Función de Distribución Acumulada (CDF) para una distribución normal. La vamos a usar para calcular el valor de la variable aleatoria bajo la distribución normal para el cual la probabilidad acumulada hasta ese punto es 1−confianza.

- Otra metodología para determinar este cálculo implica la utilización de la distribución normal estándar, donde la media (μ) es igual a 0 y la desviación estándar (σ) es 1, junto con el valor Z. Este último se puede definir como la cantidad de desviaciones estándar a las que un determinado punto se sitúa de la media. Dado que el cálculo del Valor en Riesgo (VaR) se enfoca exclusivamente en las pérdidas, adoptando así un enfoque unilateral, los valores de Z que comúnmente se emplean son:
    - 95% implica $z = 1,65$
    - 99% implica $z = 2,33$

### VaR por simulaciones de Montecarlo


Utilizamos la propuesta de Black-Scholes-Merton, para modelizar el comportamiento de un activo financiero
$$S_T=S_0e^{\left(\left(r-\frac{1}{2}\sigma^2\right)T+\sigma\sqrt{T}\mathit{z}\right)}$$
- $S_T$ valor del activo en el momento $T$
- $r$ rentabilidad libre de riesgo
- $\sigma$ desviación típica del activo
- $\mathit{z}$ variable aleatoria $N(0,1)$

- `S0 = 100` Valor inicial del activo financiero
- `r = 0.05` Rentabilidad libre de riesgo
- `sigma = 0.25`Desviación típica del activo
- `T = 2.0` Tiempo transcurrido en fracciones de año. En este caso son dos años
- `I = 10000` Número de simulaciones.
- `ST1 = S0 * np.exp((r - 0.5 * sigma ** 2) * T +sigma * np.sqrt(T) * np.random.randn(I))` replicamos la formula anterior, pero con un matiz  `np.random.randn(I)` genera mil números aleatorios que se comportan como una $N(0,1)$

```python
# Parámetros de la simulación del modelo geométrico Browniano
precio_inicial = 100   # Precio inicial del activo
rentabilidad_esperada = 0.05  # Retornos esperados
volatilidad = 0.25     # Volatilidad anual del activo
tiempo = 2.0           # Duración de la simulación en años
num_simulaciones = 100000  # Número de trayectorias simuladas

# Generar los precios finales de las simulaciones usando la fórmula del modelo geométrico Browniano
precios_finales = precio_inicial * np.exp(
    (rentabilidad_esperada - 0.5 * volatilidad ** 2) * tiempo +
    volatilidad * np.sqrt(tiempo) * np.random.randn(num_simulaciones)
)
```

Represantamos gráficamente el histográma de los posibles precios que puede tener el activo dentro de dos años.

Vamos a dibujar el histograma de los precios simulados junto con la función de distribución de probabilidad de una lognormal con:

- media $\left(r-\frac{1}{2}\sigma^2\right)T$
  
- desviación típica $\sigma *\sqrt T$.


```python
# Configurar la figura
plt.figure(figsize=(12, 6))
plt.hist(precios_finales, bins=100, density=True, alpha=0.6, color='blue', label='Histograma de precios finales')

# Calcular la media y la desviación estándar del logaritmo de los datos
mu = np.log(precio_inicial) + (rentabilidad_esperada - 0.5 * volatilidad ** 2) * tiempo
sigma_adj = volatilidad * np.sqrt(tiempo)

# Generar datos para la curva teórica de la distribución lognormal
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 200)
pdf = stats.lognorm.pdf(x, s=sigma_adj, scale=np.exp(mu))

# Dibujar la curva de la distribución lognormal teórica
plt.plot(x, pdf, 'r', linewidth=2, label='Curva teórica lognormal')

# Añadir título y leyendas
plt.title('Distribución del Precio Final del Activo Según Black-Scholes', fontsize=16)
plt.xlabel('Precio del Activo ($)', fontsize=14)
plt.ylabel('Densidad de Probabilidad', fontsize=14)
plt.legend()

# Mostrar la gráfica
plt.show()
```

# Procesos estocásticos

Un proceso estocástico es una secuencia de variables aleatorias. Podríamos verlo como una secuencia de simulaciones de variables aleatorias. Los resultados no son independientes entre si, pero fundamentalmente dependen del resultado previo. Los procesos estocásticos utilizados en finanzas siguen la regla de Markov, que indica que el precio de mañana sólo depende del estado del proceso hoy y no de los precios históricos.  Es un proceso sin memoria. Lo cúal teóricamente hablando es esencial para garantizar la eficiencia de mercado.

## Proceso Geométrico Browniano

$Z_t$ es un preceso Browniano standard o un proceso de Wiener, nombrado así en honor a Robert Brown y Norbert Wiener, es un modelo matemático utilizado para describir la evolución aleatoria de un sistema a lo largo del tiempo. Las características clave de un proceso browniano estándar $Z_t$, donde $t$ representa el tiempo, son las siguientes:

1. **Inicio en cero**: $Z_0=0$. El proceso comienza en el origen en $t =0$.

2. **Incrementos independientes**: Los incrementos del proceso en intervalos de tiempo son independientes. Es decir, si $t_1 < t_2 < t_3 < t_4$, entonces $Z_{t_2} - Z_{t_1}$ es independiente de $Z_{t_4} - Z_{t_3}$.

3. **Incrementos estacionarios y normales**: Los incrementos $Z_{t+s} - Z_s$ se comportan como una distribución normal $N(0, t)$. 

4. **Trayectorias continuas**: Casi todas las trayectorias de un proceso browniano son continuas en el tiempo. Esto significa que no hay saltos o discontinuidades en la evolución del proceso.

La ecuación diferencial que define un proceso geométrico Browiano de una acción es la siguiente:

$$dS_t=rS_tdt+\sigma S_t dZ_t$$

Como consecuencia de ello los valores de $S_t$ siguen una distribución log-normal y su rentabilidad marginal $\frac{dS_t}{S_t}$ se comporta como una normal.

Si aplicamos el esquema de Euler para pasar de tiempo continuo a tiempo discreto llegamos a la expresión de Black-Schole-Merton
$$ S_t=S_{t-\bigtriangleup t}exp\left[\left(r-\frac{1}{2}\sigma^2\right)\bigtriangleup t+\sigma\sqrt{\bigtriangleup t}z_t\right] $$

Para realizar la simulación de las sendas voy a crearme `S = np.zeros((M + 1, I))` un matriz de $M$ columnas una por cada intervalo de tiempo e $I$ filas.

```python
# Parámetros de la simulación
num_simulaciones = 100000  # Número de trayectorias simuladas
num_intervalos = 50        # Número de intervalos de tiempo
T = 1                      # Tiempo total en años
dt = T / num_intervalos    # Longitud de cada intervalo en años
precio_inicial = 100       # Valor inicial del activo
retorno_medio = 0.05   # Rentabilidad media del activo
volatilidad = 0.2          # Volatilidad anual del activo

# Inicialización de la matriz de precios del activo
precios_mc = np.zeros((num_intervalos + 1, num_simulaciones))
precios_mc[0] = precio_inicial  # Establecer el valor inicial para todas las simulaciones

# Generación de valores aleatorios para el componente de difusión
valores_aleatorios = np.random.standard_normal((num_intervalos, num_simulaciones))

# Simulación de precios utilizando operaciones vectorizadas
precios_mc[1:] = precio_inicial * np.exp(np.cumsum((retorno_medio - 0.5 * volatilidad**2) * dt 
                                                + volatilidad * np.sqrt(dt) * valores_aleatorios, axis=0))

```

El principal problema del Valor en Riesgo (VaR) es que no considera la distribución de las pérdidas que superan el umbral del VaR, es decir, no ofrece información sobre la magnitud de las pérdidas extremas.
Soluciones propuestas incluyen:
*  La **extensión de Cornish-Fisher**, que ajusta el VaR para tener en cuenta el sesgo y la curtosis de la distribución de retornos
  
*  Otras medidas como el **CVaR** (Valor en Riesgo Condicional) o **EVaR** (Valor en Riesgo Esperado) Estas medidas ofrecen una visión más completa al calcular el promedio de las pérdidas que exceden el VaR, es decir, consideran lo que sucede en la cola de la distribución de pérdidas más allá del punto del VaR



### Extensión de Cornish-Fisher del VaR Parmétrico [paper original](https://digital.library.adelaide.edu.au/dspace/bitstream/2440/15229/1/148.pdf)

La expansión de Cornish-Fisher es un método matemático usado para ajustar el cálculo del Valor en Riesgo (VaR) tomando en cuenta la asimetría (sesgo) y la curtosis (medida de colas pesadas o ligeras) de la distribución de retornos de un activo o cartera. Tradicionalmente, el VaR asume una distribución normal de los retornos, lo que puede no ser realista para muchos activos financieros que muestran comportamientos de riesgo más complejos.

El cálculo del Valor en Riesgo (VaR) utilizando la expansión de Cornish-Fisher modifica la z-score normal estándar para ajustarse a la asimetría (sesgo) y kurtosis (curtosis) de la distribución de rendimientos. La idea es corregir la aproximación normal estándar del VaR para reflejar la forma real de la distribución de los rendimientos, especialmente en las colas.

Aquí está la fórmula matemática para calcular el z-score ajustado:

$$ z_{\text{ajustado}} = z + \frac{1}{6}(z^2 - 1)S + \frac{1}{24}(z^3 - 3z)K - \frac{1}{36}(2z^3 - 5z)S^2 $$

donde:
- $ z $ es el z-score normal estándar correspondiente al nivel de confianza del VaR (por ejemplo, -1.645 para el 95% de confianza).
- $ S $ es el coeficiente de asimetría de la distribución de rendimientos.
- $ K $ es el exceso de kurtosis de la distribución de rendimientos.

Una vez calculado el $ z_{\text{ajustado}} $, se puede utilizar para calcular el VaR ajustado por Cornish-Fisher como sigue:

$$ \text{VaR}_{\text{ajustado}} = \mu + z_{\text{ajustado}} \sigma $$

donde:
- $ \mu $ es el promedio de los rendimientos.
- $ \sigma $ es la desviación estándar de los rendimientos.

Este método proporciona un VaR que tiene en cuenta la forma real de la distribución de los rendimientos, ofreciendo una medida más precisa del riesgo en distribuciones no normales.

## Maximum Drawdown

El drawdown es una medida de riesgo que refleja la máxima caída desde un pico hasta un valle en el valor de una inversión antes de alcanzar un nuevo máximo.

Esta métrica es esencial para entender la volatilidad y el riesgo real asociado a una estrategia de inversión.

A continuación, vamos a explorar algunos ejemplos de condiciones comerciales que se pueden establecer en relación con el drawdown, ilustrando su importancia en la gestión del riesgo:

* Es improbable la aceptación de un drawdown del 50%.
* Criterio de terminación: drawdown del 20%.
* Umbral de alerta: drawdown del 15%.
* Máximo período tolerado bajo el nivel de referencia: 1 año.

Max Drawdown es una medida del riesgo en las finanzas que nos ayuda a entender cuánto se puede perder en una inversión o cartera en el peor momento antes de que empiece a recuperarse. Para calcularlo:

1. **Encuentra el punto más alto**: Identifica el valor máximo alcanzado por la inversión hasta un cierto instante de tiempo.

2. **Encuentra el punto más bajo mientras no se supere el máximo**: Desde el punto más alto encontrado en el paso 1, busca el valor más bajo que alcanzó la inversión mientras no se haya superado ese máximo. Este valor más bajo representa la mayor caída o drawdown.

3. **Calcula el drawdown máximo**: Resta el valor más bajo del punto más alto encontrado en el paso 1. El resultado es el drawdown máximo

4. **Reinicia el cálculo después de superar el punto más alto**: Después de que el valor de la inversión supere el punto más alto anterior (el máximo registrado), comienza un nuevo cálculo de drawdown máximo. Esto significa que cualquier pérdida anterior se considera parte del drawdown pasado, y se inicia un nuevo registro del drawdown a partir del nuevo punto más alto.

El drawdown máximo evoluciona a medida que la inversión alcanza nuevos máximos. 

El drawdown máximo de cada activo se calcula siguiendo la formula:

$$Drawdown = \frac{P_t - P_{max}}{P_{max}} = \frac{P_t}{P_{max}}-1$$

```python
pesos = cp.Variable(num_act)
t = cp.Variable()
uj = cp.Variable(num_data)
dj = cp.Variable(num_data+1)

riesgo = t + cp.sum(uj)/(alpha * num_data)
retornos = retornos_esperados.T @ pesos
```

```python
restricciones = [dj[1:] - t - uj <= 0,
                dj[1:] >= dj[:-1] - ret @ pesos,
                uj >= 0,
                dj >= 0,
                dj[0] == 0,
                cp.sum(pesos) == 1,
                pesos >= 0,
#                retornos >= 0.0007,
            ]
```

```python
objective = cp.Minimize(riesgo)

# Solve the problem
prob = cp.Problem(objective, restricciones)
prob.solve()
```

## Revisión de la Metodología: Evitando el Sesgo de Anticipación

En el ejercicio anterior, utilizamos **los mismos datos para calcular los pesos de las carteras y para evaluar su evolución**. Este enfoque puede introducir un **sesgo de look-ahead**, donde se utiliza información futura al momento de tomar decisiones de inversión, lo cual no es posible en un entorno real de trading.

### Sesgo de Look-Ahead

El sesgo de look-ahead ocurre cuando se utilizan datos que no estarían disponibles en el momento de la toma de decisiones en el mundo real. Esto puede llevar a resultados de rendimiento inflados y no realistas.

### Nuevo Enfoque: División de Datos en Train y Test

Para evitar este sesgo y mejorar la robustez de nuestras estrategias de inversión, vamos a modificar nuestra metodología:

1. **División de Datos**: Separaremos nuestro conjunto de datos en dos partes:
   - **Datos de Entrenamiento (Train)**: Usados para calcular los pesos de las carteras.
   - **Datos de Prueba (Test)**: Usados para evaluar la evolución de las carteras.

2. **Cálculo de Pesos en el Conjunto de Train**:
   - Aplicaremos los criterios de selección de carteras (mínima volatilidad, máximo ratio de Sharpe, mínimo CVaR y cartera equitativa) solo a los datos de entrenamiento.

3. **Evaluación de los Retornos en el Conjunto de Test**:
   - Una vez definidos los pesos de las carteras, utilizaremos los datos de prueba para evaluar cómo habrían actuado estas carteras en el mercado, sin el sesgo de tener acceso a esta información durante la fase de cálculo de los pesos.

### Implementación

Este enfoque no solo nos proporciona una evaluación más precisa y realista de las estrategias de cartera, sino que también nos ayuda a comprender mejor cómo podrían comportarse estas carteras bajo diferentes condiciones de mercado que no fueron consideradas durante la fase de optimización.
